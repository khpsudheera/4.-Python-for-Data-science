```python
#import libraries
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read file
Car_data = pd.read_excel("C:/Users/91918/OneDrive/Desktop/SUDHEERA/DataScience with Gen Ai/Course4-Numpy python/Project/C4_Project_data.xlsx")
Car_data
```

```python
#Task1:
#ABS Analysis
abs_count = Car_data['ABS_(Anti-lock_Braking_System)'].notna().sum()
abs_total = len(Car_data)
abs_percentage = (abs_count / abs_total) * 100
no_abs_count = abs_total - abs_count
print(f"Models with ABS: {abs_count} ({abs_percentage:.2f}%)")
print(f"Models without ABS: {no_abs_count} ({100 - abs_percentage:.2f}%)")
#Airbags analysis
airbag_types = Car_data['Airbags'].dropna().unique()
airbag_count = Car_data['Airbags'].notna().sum()
print(f"Models with Airbags: {airbag_count}")
print(f"Types of Airbags: {', '.join(airbag_types)}")
#hill assis analysis
hill_assist_count = Car_data['Hill_Assist'].notna().sum()
hill_assist_total = len(Car_data)
hill_assist_percentage = (hill_assist_count / hill_assist_total) * 100
print(f"Models with Hill Assist: {hill_assist_count} ({hill_assist_percentage:.2f}%)")
print(f"Models without Hill Assist: {hill_assist_total - hill_assist_count} ({100 - hill_assist_percentage:.2f}%)")
#Interpretation:here model with ABS is 546 (87.08%) and without ABS is 81 (12.34%).
#and here many types in airbags and models with arbags is 540
```

```python
#Task2:
seating_capacity_col = 'Seating_Capacity'
central_locking_col = 'Central_Locking'
child_safety_locks_col = 'Child_Safety_Locks'

# Average Seating Capacity
average_seating_capacity = Car_data[seating_capacity_col].mean()
print(f"Average Seating Capacity: {average_seating_capacity:.2f}")

# Central Locking
central_locking_count = Car_data[central_locking_col].str.contains('Yes', na=False).sum()
central_locking_percentage = (central_locking_count / len(Car_data)) * 100
print(f"Models with Central Locking: {central_locking_count} ({central_locking_percentage:.2f}%)")

# Child Safety Locks
child_safety_locks_count = Car_data[child_safety_locks_col].str.contains('Yes', na=False).sum()
child_safety_locks_percentage = (child_safety_locks_count / len(Car_data)) * 100
print(f"Models with Child Safety Locks: {child_safety_locks_count} ({child_safety_locks_percentage:.2f}%)")
```

```python
#Interpretation:here Average Seating Capacity: 5.33.
#Models with Central Locking: 551(87.88%).
#Models with Child Safety Locks: 594 (94.74%)
```

```python
#Task3:
#Count Models with High-Speed Alerts:
high_speed_alert_count = Car_data['High_Speed_Alert_System'].str.contains('Yes', na=False).sum()
total_models = len(Car_data)
high_speed_alert_percentage = (high_speed_alert_count / total_models) * 100
print(f"Models with High-Speed Alerts: {high_speed_alert_count} ({high_speed_alert_percentage:.2f}%)")
#Count models with Seat Belt Reminders
seat_belt_reminder_count = Car_data['Fasten_Seat_Belt_Warning'].str.contains('Yes', na=False).sum()
seat_belt_reminder_percentage = (seat_belt_reminder_count / total_models) * 100
print(f"Models with Seat Belt Reminders: {seat_belt_reminder_count} ({seat_belt_reminder_percentage:.2f}%)")
#Count Models with Door-Ajar Warnings:
door_ajar_warning_count = Car_data['Door_Ajar_Warning'].str.contains('Yes', na=False).sum()
door_ajar_warning_percentage = (door_ajar_warning_count / total_models) * 100
print(f"Models with Door-Ajar Warnings: {door_ajar_warning_count} ({door_ajar_warning_percentage:.2f}%)")
#Interpretation:Models with Seat Belt Reminders: 540 (86.12%),here many models have Door-Ajar Warnings that is 567(90.43%) and models have high speed alert system that is 42(6.70%)
```

```python
#Task4:

height_col = 'Height_mm'
length_col = 'Length_mm'
width_col = 'Width_mm'

# Summary statistics for each dimension
height_stats = Car_data[height_col].describe()
length_stats = Car_data[length_col].describe()
width_stats = Car_data[width_col].describe()

print("Height Statistics:")
print(height_stats)
print("\nLength Statistics:")
print(length_stats)
print("\nWidth Statistics:")
print(width_stats)
make_wise_averages = Car_data.groupby('Make')[[height_col, length_col, width_col]].mean().reset_index()
print("Make-Wise Average Dimensions:")
print(make_wise_averages)
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")

# Plotting average dimensions by make
plt.figure(figsize=(14, 8))

# Plot Height
plt.subplot(3, 1, 1)
sns.barplot(x='Make', y=height_col, data=make_wise_averages, palette='viridis')
plt.title('Average Height by Make')
plt.xticks(rotation=90)

# Plot Length
plt.subplot(3, 1, 2)
sns.barplot(x='Make', y=length_col, data=make_wise_averages, palette='viridis')
plt.title('Average Length by Make')
plt.xticks(rotation=90)

# Plot Width
plt.subplot(3, 1, 3)
sns.barplot(x='Make', y=width_col, data=make_wise_averages, palette='viridis')
plt.title('Average Width by Make')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
#Interpretation:The analysis reveals that car manufacturers vary significantly in their preferred dimensions. While some prioritize larger vehicles (e.g., Aston Martin, Audi, Bentley), others focus on smaller, more maneuverable models (e.g., Datsun, Fiat, Force). These findings can inform product development and marketing strategies for different target markets.
```
