```python
#task2
# Define the driving modes to check for
required_modes = {'Normal', 'Comfort', 'Eco', 'Sport', 'Power'}

# Function to check if a car has all required driving modes
def has_all_modes(modes_str):
    if pd.isna(modes_str):
        return False
    modes = set(mode.strip() for mode in modes_str.split(','))
    return required_modes.issubset(modes)

# Filter cars that have all the required driving modes
filtered_cars = Car_data[Car_data['Drive_Modes'].apply(has_all_modes)]

# Display the details of these cars
print(filtered_cars)

# Count how many cars have all the required driving modes
num_cars_with_all_modes = filtered_cars.shape[0]
print(f"Number of cars with all the specified driving modes: {num_cars_with_all_modes}")
```

```python
#Task3:
# Convert the city mileage column to numeric (handle any non-numeric values or errors)
Car_data['City_Mileage_km_litre'] = pd.to_numeric(Car_data['City_Mileage_km_litre'], errors='coerce')

# Group by manufacturer and calculate the mean city mileage
manufacturer_mileage = Car_data.groupby('Make')['City_Mileage_km_litre'].mean()

# Get the top 10 manufacturers based on average city mileage
top_10_manufacturers = manufacturer_mileage.nlargest(10)

# Display the results
print(top_10_manufacturers)

# Plot the top 10 manufacturers using a horizontal bar chart
plt.figure(figsize=(12, 8))
top_10_manufacturers.plot(kind='barh', color='skyblue')
plt.xlabel('Average City_Mileage_km_litre')
plt.ylabel('Manufacturer')
plt.title('Top 10 Car Manufacturers Based on City Mileage')
plt.gca().invert_yaxis()  # To display the highest mileage at the top
plt.show()

# Check for outliers in city mileage
mileage_stats = Car_data['City_Mileage_km_litre'].describe()
print(mileage_stats)

# Plot the distribution of city mileage
plt.figure(figsize=(10, 6))
plt.hist(Car_data['City_Mileage_km_litre'].dropna(), bins=30, color='lightgreen', edgecolor='black')
plt.xlabel('City_Mileage_km_litre')
plt.ylabel('Frequency')
plt.title('Distribution of City Mileage')
plt.show()
```

```python
#task3(continution):
# Removing outliers using IQR method
Q1 = Car_data['City_Mileage_km_litre'].quantile(0.25)
Q3 = Car_data['City_Mileage_km_litre'].quantile(0.75)
IQR = Q3 - Q1

# Define acceptable range for mileage
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
filtered_df = Car_data[(Car_data['City_Mileage_km_litre'] >= lower_bound) & (Car_data['City_Mileage_km_litre'] <= upper_bound)]

# Re-calculate and plot the top 10 manufacturers
manufacturer_mileage_filtered = filtered_df.groupby('Make')['City_Mileage_km_litre'].mean()
top_10_manufacturers_filtered = manufacturer_mileage_filtered.nlargest(10)

# Plot the results
plt.figure(figsize=(12, 8))
top_10_manufacturers_filtered.plot(kind='barh', color='skyblue')
plt.xlabel('Average City_Mileage_km_litre')
plt.ylabel('Manufacturer')
plt.title('Top 10 Car Manufacturers Based on City Mileage (Filtered)')
plt.gca().invert_yaxis()
plt.show()
```
